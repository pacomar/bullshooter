```python
url2 = 'https://www.bullshooter.eu/european-couples-league-2023/'
response2 = requests.get(url2, headers=headers)
soup2 = BeautifulSoup(response2.content, "html.parser")
```

```python
def parsear_tabla_html(tabla):    
    # Obtén las filas de la tabla
    filas = tabla.find_all('tr')
    
    # Obtén los encabezados de la tabla
    encabezados = [encabezado.text.strip() for encabezado in filas[0].find_all('td')]
    
    # Parsea los datos de la tabla en una lista de diccionarios
    datos = []
    for fila in filas[1:]:
        celdas = fila.find_all('td')
        fila_datos = {}
        for i, celda in enumerate(celdas):
            fila_datos[encabezados[i]] = celda.text.strip()
        datos.append(fila_datos)
    
    return datos
```

```python
groups = []
groups_html = soup2.find_all("div", class_="elementor-tab-title")
for group in groups_html:
    name = group.find_all('a')[0].text
    groups.append({'name': name})
tables = soup2.find_all("table")
for i, table in enumerate(tables):
    group = parsear_tabla_html(table)
    groups[i]['table'] = group
elementos_p = soup2.find_all('p')
hrefs = []
for elemento_p in elementos_p:
    href = elemento_p.find('a')
    if href != None: hrefs.append(href)
for i, href in enumerate(hrefs):
    groups[i]['href'] = href.get('href') if href else None    
groups
```

```python
for group in groups:
    responseScore = requests.get(group['href'])
    soupScore = BeautifulSoup(responseScore.content, "html.parser")
    print(soupScore)
```

```python
def find_index(list, text_to_find):
    for i, item in enumerate(list):
        if text_to_find in item:
            return i
    return -1
```
